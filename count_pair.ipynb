```python
import bisect
#bisect is based on bisecting algorithm which helps us to find the index of elelment which when inserted will
# done so that reesulatant array remains in sorted and time complexity is O(log(n)) for this.
class solution:
    # function to find the unique pair so that sum is equal to k 
    def count_array(self,arr,k):
        # let us sort the array
        arr.sort()
        #initializing count and x equal to zer
        count=0
        x=0
        # len of the input array
        n=len(arr)
        #[1,1,5,7]
        # iterating in a loop till the 2nd last element
        for i in range(len(arr)-1):
            # find the number after subtracting the a[i] elements
            x=k-arr[i]
            #now let us find at what index this x to be  left inserted so arrays is sorted and here the lo=i+1 hi =n which means it should be inserted between this index
            y=bisect.bisect_right(arr,x,i+1,n)
            """now let us find at what index this x to be  left inserted so arrays is sorted .
            element is inserted at left to make array sorted but range of selection is from i+1 to n
             as our whole goal is to find unique pairs where sum is equal to k as we move towards right the 
             value at x starts decreasing and from biset function we have lower range value is i+1 which means that it can not be inserted 
             at the right value so the value of uppperbound and lowerbound will be same and it will cancel out 
             due to which we dont get repeated pairs"""
            z=bisect.bisect_left(arr,x,i+1,n)
            count=count+bisect.bisect_right(arr,x,i+1,n)-bisect.bisect_left(arr,x,i+1,n)
        # return the count   
        return count
            
# here we use time complecity of O(n*log(n)) and space complexity is O(1)
        
        
        
c1=solution()
c1.count_array([1, 5, 7, 1],6)

            

```

```python
# we will use hashing to tackle the same problem
class solution:
    # defining the function to find the count of pairs whose sum is equal.
    def sum_count(self,arr,k):
        # sorting the array
        arr.sort()
        # creating a different array to store the frequency
        m=[0]*1000
        """ iterating through the array to store the elements in form of key and value where key is element 
        at index and number is its frequency for e.g [1,1,3,5] as [0,2(index of 2 and count is 2),0,1,0,1] """
        for i in range(len(arr)):
            m[arr[i]]+=1
        # initializing the double count
        double_count=0
        # iterating in the loop in array
        for i in range(len(arr)):
            # iterating in the loop and finding the frequency of the required number and adding it to the count
            double_count+=m[k-arr[i]]
            # to remove the double iteration in the loop
            if m[k-arr[i]]==arr[i]:
                # remove the value from the count in the double_count
                double_count-=1
                # return the value
        return double_count
            
            
# now we have to use hashing
        
        
        
c1=solution()
c1.sum_count([1, 5, 7, 1],6)
            


            
```
